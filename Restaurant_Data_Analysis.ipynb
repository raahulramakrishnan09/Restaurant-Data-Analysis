```python
#Finding_missing_values
import missingno as msno
msno.bar(df)
plt.show()
```

```python
#Dropping_the_table_with_more_missing_values
df=df.drop(['Date','HSN','Timestamp','Covers','Table No.','Server Name','Variation'],axis=1)
```

```python
#Total_orders
avg1=df['Invoice No.'].nunique()
```

```python
#Total_amount
avg2=df['Final Total'].sum()
```

```python
#Average_order_value
avg=avg2/avg1
avg
```

```python
#Using_Groupby
new_df=df.groupby('Invoice No.').agg({'Invoice No.': lambda x:x.nunique(),'Qty.': lambda x:x.sum(),
                                      'Final Total': lambda x:x.sum(),'Item Name': lambda x:x.sum(),})
new_df.columns=['INVOICE_NO','QUANTITY','AMOUNT','ITEM_NAME']
new_df
```

```python
#visualizing the top_10 Items using barplot
top_10_descriptions=df['Item Name'].value_counts().nlargest(10)
sns.barplot(x=top_10_descriptions.values,y=top_10_descriptions.index)
plt.xlabel('Count')
plt.ylabel('Items')
plt.title('Top 10 Items Descriptions')
plt.show()
```

```python
#visualizing the top_10 categories using barplot
top_10_descriptions=df['Category'].value_counts().nlargest(10)
sns.barplot(x=top_10_descriptions.values,y=top_10_descriptions.index)
plt.xlabel('Count')
plt.ylabel('categories')
plt.title('Top 10 Category Descriptions')
plt.show()
```

```python
#visualizing the top_10 Items based on Quantity
top_10=df.groupby('Item Name')['Qty.'].sum().nlargest(10)
sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Quantity')
plt.ylabel('Items')
plt.title('Top 10 Items based on Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
#visualizing the top_10 Items based on Final Total
top_10=df.groupby('Item Name')['Final Total'].sum().nlargest(10)
sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Amount')
plt.ylabel('Items')
plt.title('Items based on Final Total')
plt.show()
```

```python
#visualizing the top_05 Category based on Discount
top_10=df.groupby('Category')['Discount'].mean().nlargest(5)
sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Total Discount')
plt.ylabel('Category based on Discount')
plt.title('Items')
plt.show()
```

```python
#visualizing the top_05 Category based on Quantity
top_10=df.groupby('Category')['Qty.'].sum().nlargest(5)
sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Quantity')
plt.ylabel('Category')
plt.title('Category based on Quantity')
plt.show()
```

```python
#visualizing the top_05 Category based on Final Total
top_10=df.groupby('Category')['Final Total'].sum().nlargest(5)
sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Amount')
plt.ylabel('Category')
plt.title("Category based on Final Total")
plt.show()
```

```python
#visualizing the top_10 Invoice No. and Items together based on Quantity
top_10=df.groupby(['Invoice No.','Item Name'])['Qty.'].sum().nlargest(10)
top_10.plot(kind='bar',color='skyblue')
```

```python
#Number_of_Fish_65_Ordered
k=df[df['Item Name']=='Fish 65']
print("Number of Fish_65 Ordered:",k['Qty.'].sum())
```

```python
#Number_of_Customers_Ordered_Parota
p=df[df['Item Name']=='Parota']
p1=p.groupby('Item Name').agg({'Invoice No.': lambda x:x.nunique()})
p1.columns=['NO_OF_CUSTOMERS_ORDERED']
p1
```

```python
#Quantity_of_Blue_Lagoon
m=df[df['Item Name']=='Blue Lagoon']
m1=m.groupby('Item Name').agg({'Qty.':lambda x:x.sum()})
m1.columns=['NUMBER_OF_ITEMS_ORDERED']
m1
```

```python
#Number_of_Customers_Ordered_and_Quantity_of_Romali_Roti(1 No)
n=df[df['Item Name']=='Romali Roti (1 No)']
n1=n.groupby('Item Name').agg({'Qty.':lambda x:x.sum(),
                               'Invoice No.': lambda x:x.nunique()})
n1.columns=['NUMBER_OF_ITEMS_ORDERED','NO_OF_CUSTOMERS_ORDERED']
n1
```

```python
#Number_of_Invoices_with_charged_tax_of_14
o=df[df['Tax']==14]
o1=o.groupby('Tax').agg({'Invoice No.':lambda x:x.nunique()})
o1.columns=['NO_OF_CUSTOMERS_ORDERED']
o1
```

```python
#Number_of_'Plain Briyani With Raitha, Gravy.'_ordered_with_Discount_of_6.5
q=df[(df['Item Name']=='Plain Briyani With Raitha, Gravy.') & (df['Discount']==6.5)]
print("Number of 'Plain Briyani With Raitha, Gravy.' ordered with Discount of 6.5:",q['Invoice No.'].nunique())
```
